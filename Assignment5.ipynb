파이썬 코드가 다형적이라는 의미는 여러가지 형에 대해 동작할 수 있는 코드임을 말한다. 아래 코드에서는 s와 ss가 문자열과 리스트로 정의되어있다. 그리고 count_words라는 함수에서 s와 ss내에 존재하는 글자의 빈도수를 정확히 계산한 모습니다. 코드의 재사용을 촉진하는 다형성의 의미에 비추어 봤을 때, 아래 count_words()함수는 리스트건 문자열이건 딕셔너리에서도 동작하는 다형적인 코드이다.

```python
def count_words(s):
    dic = dict()
    for i in s:
        if i not in dic:
            dic[i] = 1
        else:
            dic[i] = dic[i] + 1
    return dic

s = "123123123"
ss = ['a', 'a', 'b']
print count_words(s)
print count_words(ss)
```

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def incr(self):
        self.count += self.step
        return self.count
    def __call__(self):
        self.incr()
        return self.count
    def __add__(self, i):
        self.count = self.count + i
        return self
    def __sub__(self, i):
        self.count = self.count - i
        return self
    def __str__(self):
        return "[Count (step: " + str(self.step) + ")] "+ str(self.count)
    def __lt__(self, y):
        return int(self.count) < y
    def __gt__(self, y):
        return int(self.count) > y
    def __eq__(self, y):
        return int(self.count) == y
    def __ne__(self, y):
        return int(self.count) != y
c = Counter(10)
d = Counter(10, 2)
print c
print d

c.incr()
d.incr()
print c
print d

c()
d()
print c
print d

c = c + 5
d = d - 5
print c
print d

print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

```python
#4번 문제
#다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 
#다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 
#내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
```

__init__메소드는 객체가 생성될 때, 자동으로 불려지는 메소드이다. __init__메소드는 list를 인자로 받아 객체 인스턴스에 list의 원소들을 append하고, 객체 인스턴스를 매개로 하여, 정적메소드 eliminate_duplicate()를 부르고 있다.

여기서 eliminate_duplicate()메소드는 객체 인스턴스를 받아 객체 인스턴스의 중복을 제거하여 다시 객체 인스턴스에 넣어주고 있다. 정적메소드는 클래스 이름을 통해 직접 부르고 있어, 객체 인스턴스를 통해서도 호출이 가능하기에 객체 인스턴스의 중복을 제거하는 코드가 가능토록 하고있다.

다음으로 __str__메소드는 객체를 출력하기 위한 print 예약어 호출에 대응이 된다. 즉 "print 객체"를 통해서 불려지는 메소드이다. 여기서는 return값으로 {객체의 원소들, ... }을 주고 있어, 중복이 제거된 객체 인스턴스의 원소들을 중괄호 내에 추가하여 return하고 있는 모습이다.

```python
#5번 문제
#4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오
```

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
    def __or__(self, B): #합집합 연산
        for e in B: #B의 원소들이 객체 인스턴스에 존재하지 않는다면 append
            if e not in self:
                self.append(e)
        return self #return 객체인스턴스
    def __and__(self, B): #교집합 연산
        l = []
        for e in B: #B의 원소들이 객체 인스턴스에 존재한다면 임의의 리스트 l에 삽입
            if e in self:
                l.append(e)
        self[:] = l #객체 인스턴스를 리스트 l로 바꾸어주어 return한다.
        return self
    def __sub__(self, B): #차집합 연산
        for e in B:#B의 원소들이 객체 인스턴스에 존재한다면 객체 인스턴스 내에서 해당 원소 제거
            if e in self:
                self.remove(e)
        return self

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t
    #u = s | t
    #print u
    u = s & t 
    print u
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t 
    print u 
```

```python
#10번 문제 incremental project
import urllib2
import string
import json
class WebWordsFrequency():
    def __init__(self, *url):
        self.url = list()
        for e in url:
            self.url.append(e)
    def addUrl(self, url):
        self.url.append(url)
    def removeUrl(self, url):
        self.url.remove(url)
    def listUrls(self):
        for i in self.url:
            print i
    def getDict(self, url):
        L_source = list()
        for e in url:
            L_source.append(urllib2.urlopen(e).read())
        lock = False
        comment = False
        end_comment = 0
        for i in range(len(L_source)):
            if L_source[i] == '<':
                lock = True
                end_comment -= 1
                L_source[i] = ''
            if L_source[i] == '>':
                lock = False
                end_comment += 1
                L_source[i] = ''
            if lock == True and L_source[i] == '!':
                comment = True
                end_comment = 0
            if comment == True and end_comment == 1:
                comment = False
            if lock == True:
                L_source[i] = ''
            if comment == True:
                L_source[i] = ''
        result = "".join(L_source)
        count = 0
        text_line = result.split()
        for i in range(0, len(text_line)):
            count += len(text_line[i].split(' '))
        
        #불용어 리스트 만들기
        f = open("stop_words.txt")#이 파일안에 불용어들이 있다.
        l = f.readlines()
        f.close()
        stop_words = list()
        for i in l:
            if i == '\n':
                continue
            stop_words.append(i.strip())
        
        #순수 텍스트만 추출
        result_list = []
        for i in text_line:
            S = ''
            for j in range(len(i)):
                if i[j] not in string.punctuation:
                    S += i[j]
            if len(S) != 0 and S not in stop_words:
                result_list.append(S)
        #사전화 작업        
        count_word = dict()
        for i in result_list:
            if i not in  count_word:
                count_word[i] = 1
            else:
                count_word[i] += 1
        return count_word
    def getWordsFrequency(self):
        mydict = self.getDict(self.url)
        return mydict
        
    def getMaxFreqencyWords(self):
        frequency_dict = self.getWordsFrequency()
        
        dict_keys = frequency_dict.keys()
        dict_values = frequency_dict.values()
    
        for i in range(len(dict_values)-1):
            for j in range(len(dict_values)-i-1):
                if dict_values[j] < dict_values[j+1]:
                    dict_values[j], dict_values[j+1] = dict_values[j+1], dict_values[j]
                    dict_keys[j], dict_keys[j+1] = dict_keys[j+1], dict_keys[j]
        index = dict_values[0]
        result = list()
        result.append(dict_keys[0])
        for i in range(1, len(dict_values)):
            if index == dict_values[i]:
                index = dict_values[i]
                result.append(dict_keys[i])
            else:
                break
        return result
    def searchUrlByWord(self, word):
        similarity = dict() #유사도 사전
        for e in self.url:
            in_list = list()
            in_list.append(e)
            search_dict = self.getDict(in_list)
            similar_x = 1
            dict_keys = search_dict.keys()
            dict_values = search_dict.values()
    
            for j in range(len(dict_values)-1):
                for k in range(len(dict_values)-j-1):
                    if dict_values[k] < dict_values[k+1]:
                        dict_values[k], dict_values[k+1] = dict_values[k+1], dict_values[k]
                        dict_keys[k], dict_keys[k+1] = dict_keys[k+1], dict_keys[k]
            #단어의 빈출 수를 곱한다. 일치하면 빈출수*2를 더하고 포함되면 빈출수*1을 더한다.
            for j in range(len(dict_keys)):
                if word == dict_keys[j]:
                    similar_x += dict_values[j]*2
                elif word in dict_keys[j]:
                    similar_x += dict_values[j]
            similarity[e] = similar_x
        return min(similarity)
```

```python
#11번 문제, 12번 문제
class OrderedWebWordsFrequency(WebWordsFrequency):
    def __init__(self, *url):
        self.url = list()
        self.index = 0
        for e in url:
            self.url.append(e)
        
    def getKey(self, item):
        return item[1]
    def getWordsFrequency(self, reverse = False):
        mydict = self.getDict(self.url)
        keys = mydict.keys()
        values = mydict.values()
        self.mylist = list(zip(keys, values))
        if reverse == False:
            self.mylist=sorted(self.mylist, key = self.getKey, reverse = True)
        else:
            self.mylist=sorted(self.mylist, key = self.getKey, reverse = False)
        return self.mylist
    
    #반복자
    def __iter__(self): 
        return self
    def next(self):
        try:
            value = self.mylist[self.index]
            self.index += 1 
            return value
        except:
            raise StopIteration
```
